```python
import pandas as pd
```

```python
union_df = pd.read_html("https://maple.gg/info/union")
union_df[6]
```

```python
union_effect_df = [1,2,3,4,5,6]
for i in range(1,len(union_df)):
    union_effect_df[i-1] = pd.DataFrame(union_df[i],columns = ["직업/효과","효과"])
union_effect_df[2].head(3)
```

```python
all_union_effect_df = pd.concat([union_effect_df[0],union_effect_df[1],union_effect_df[2],union_effect_df[3],
                                 union_effect_df[4],union_effect_df[5]],ignore_index=True)
print(all_union_effect_df.head(3))
print(all_union_effect_df.tail(3))
```

```python
all_union_effect_df.rename(columns={"직업 효과":"직업"},inplace=True)
all_union_effect_df.head(3)
```

```python
all_union_effect_df["직업"] = all_union_effect_df["직업"].str.split(" ").str[0]
all_union_effect_df
```

```python
def counting_effect(df,effect):
    n, result = 0 , 0
    
    while True:
        print(df.iloc[n][0])
        if effect == df.iloc[n][1]:
            result += 1
        elif result != 0 and df.iloc[n][1] != effect:
            break
        n += 1
    
    return result
```
